## 퍼셉트론 초기 구축

```python
import numpy as np
```

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7  #theta -> 임계값
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,0)
AND(1,0)
AND(0,1)
AND(1,1)
```

#### 가중치와 편향 도입

```python
x = np.array([0,1]) #입력
w = np.array([0.5, 0.5]) #가중치
b = -0.7 # 편향
```

```python
w*x #인덱스가 같은 원소끼리 곱한다.
```

```python
np.sum(w*x) # np.sum() 메서드는 입력한 배열에 담김 모든 원소의 총합을 계산.
```

```python
np.sum(w*x) + b #대략 -0.2 (부동소수점 수에 의한 연산 오차)
```

```python
## 가중치와 편향 구현하기
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
    
# w1, w2는 각 입력 신호가 결과에 주는 영향력(중요도)
# 편향은 뉴런이 얼마나 쉽게 활성화(결과 1을 출력)하느냐를 조정하는 매개 변수이다.
```

```python
AND(0,0)
```

#### NAND 게이트와 OR 케이트

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) # AND와 가중치(w와 b)만 다르다!
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) # AND와는 가중치(w와 b)만 다르다
    b = -0.2
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

#### XOR 게이트 구현하기

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,0)
```

```python
XOR(0,1)
```

```python
XOR(1,1)
```

##### XOR 게이트는 단층 퍼셉트론으로는 표현할 수 없다.
##### 2층 퍼셉트론을 이용하면 XOR게이트를 표현할 수 있다.
##### 단층 퍼셉트론은 직선형 영역만 표현할 수 있고, 다층 퍼셉트론은 비선형 영역도 표현할 수 있다.
